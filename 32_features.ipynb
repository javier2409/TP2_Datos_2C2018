```python
import pandas as pd
```

```python
file = pd.read_csv('events_up_to_01062018.csv',low_memory=False)
```

```python
features_1 = pd.read_csv('features_10.csv')
```

```python
features_1 = features_1.set_index('person')
```

```python
precios = pd.read_csv('precios_celulares.csv')
```

```python
precios.columns=['model','precio']
```

## Promedio de precios de una region

```python
new_file = file.merge(precios,on='model',how='left')
```

```python
new_file = new_file.set_index('person').join(features_1,how='left',lsuffix='_left', rsuffix='_right')
```

```python
a=new_file.loc[new_file['event']=='viewed product'].groupby('region_sin_unknown')["precio"].mean().to_frame().reset_index()
a.columns=['region_sin_unknown','precio_visto_promedio_region']
```

```python
b=new_file.loc[new_file['event']=='conversion'].groupby('region_sin_unknown')["precio"].mean().to_frame().reset_index()
b.columns=['region_sin_unknown','precio_compras_promedio_region']
```

```python
features_1=features_1.reset_index()
```

```python
features_1= features_1.merge(a,on="region_sin_unknown",how='left')
features_1= features_1.merge(b,on="region_sin_unknown",how='left')
```

## Diferencia de precios entre el promedio de la region y el usuario

```python
features_1["diferencia_con_la_region"] = features_1['precio_visto_promedio_region']-features_1["promedio_precio_celulares_vistos"]
```

```python
features_1["diferencia_con_la_region_n"] = (features_1["diferencia_con_la_region"] - features_1["diferencia_con_la_region"].mean() ) / features_1["diferencia_con_la_region"].std()
```

```python
features_1.head()
```

```python
features_1.to_csv("features_11.csv")
```
