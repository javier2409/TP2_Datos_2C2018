```python
from scipy.stats import rv_discrete
```

```python
import pandas as pd
```

```python
file = pd.read_csv('events_up_to_01062018.csv',low_memory=False)
```

```python
features_1 = pd.read_csv('features_09.csv')
```

```python
features_1 = features_1.drop('Unnamed: 0',axis=1).set_index('person')
```

```python
precios = pd.read_csv('precios_celulares.csv')
```

```python
precios.columns=['model','precio']
```

## Promedio del precio de los celulares vistos

```python
new_file = file.merge(precios,on='model',how='left')
```

```python
new_file.loc[new_file.event=='viewed product','precio'].mean()
```

```python
def promedioprecio(x):
    a = x.loc[x.event=='viewed product','precio'].mean()
    return a
promediopreciovio = new_file.groupby('person').apply(promedioprecio)
promediopreciovio = pd.DataFrame(promediopreciovio)
promediopreciovio.columns = ['promedio_precio_celulares_vistos']
```

```python
features_1 = features_1.join(promediopreciovio,how='left')
```

## Rellenando los unknown de la columna 'region' por un valor segun la distribucion de las regiones

```python
regs = features_1.region.value_counts().drop('Unknown')
```

```python
regs_vals = list(regs.index)
regs_probs = regs.values/regs.values.sum()
```

```python
region_distrib = rv_discrete(values=(range(len(regs_vals)),regs_probs))
```

```python
region_sin_unknown = []
for i in features_1.region.values:
    if i == 'Unknown':
        region_sin_unknown.append(regs_vals[region_distrib.rvs(size=1)[0]])
    else:
        region_sin_unknown.append(i)
```

```python
features_1['region_sin_unknown'] = region_sin_unknown
```

## De que pais es la persona

```python
visitas = file.loc[file.event=='visited site']
def pais (x):
    return x.country.value_counts().index[0]
pais = visitas.groupby('person').apply(pais)
    
```

```python
pais = features_1.join(pd.DataFrame(pais),how='left')[0].fillna('Unknown')
```

```python
paises = pais.value_counts().drop('Unknown')
```

```python
pais_vals = list(paises.index)
pais_probs = paises.values/paises.values.sum()
```

```python
pais_distrib = rv_discrete(values=(range(len(pais_vals)),pais_probs))
```

```python
pais_sin_unknown = []
for i in pais.values:
    if i == 'Unknown':
        pais_sin_unknown.append(pais_vals[pais_distrib.rvs(size=1)[0]])
    else:
        pais_sin_unknown.append(i)
```

```python
features_1['pais_sin_unknown'] = pais_sin_unknown
```
